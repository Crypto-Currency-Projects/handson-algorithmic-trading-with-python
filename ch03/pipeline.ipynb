```python
# Configuring and importing the dataset
```

```python
from quantopian.research import symbols
symbols(symbols=['SPY', 'DIA', 'NDAQ'])
```

```python
from quantopian.pipeline.data.builtin import USEquityPricing
USEquityPricing()
```

```python
from quantopian.pipeline.filters import StaticAssets
etfs = (StaticAssets(symbols([
    'SPY',
    'DIA',
    'QQQ',
    'IWM',
    'EFA',
    'EEM'
])))
etfs
```

```python
from quantopian.pipeline.filters import QTradableStocksUS
us_tradeable = QTradableStocksUS()

from quantopian.pipeline import Pipeline
pipe_us_tradeable = Pipeline(screen=us_tradeable)
us_tradeable, pipe_us_tradeable
```

```python
spy_etf = (StaticAssets(symbols([
    'SPY'
])))
spy_etf
```

```python
# Defining the computations and filters
```

```python
from quantopian.research import run_pipeline
pipe_us_tradeable = Pipeline(screen=us_tradeable)
pipe_us_tradeable
```

```python
pipe_etfs_volume = Pipeline(columns={'volume': USEquityPricing.volume.latest}, screen=etfs)
pipe_etfs_volume
```

```python
pipe_spy = Pipeline(columns={'price': USEquityPricing.close.latest}, screen=spy_etf)
pipe_spy
```

```python
# Initializing and running the pipeline
```

```python
results_us_tradeable = run_pipeline(pipe_us_tradeable, start_date='2020-01-01', end_date='2020-01-01')
results_us_tradeable
```

```python
results_etfs_volume = run_pipeline(pipe_etfs_volume, start_date='2020-01-01', end_date='2020-01-01')
results_etfs_volume
```

```python
# Components of a Pipeline
```

```python
# Factors
```

```python
from quantopian.pipeline.factors import BollingerBands
lower_band, middle, upper_band = BollingerBands(window_length=22, k=2)
lower_band, middle, upper_band
```

```python
pipe_spy.add(lower_band, 'lower_band', overwrite=True)
pipe_spy.add(middle, 'middle', overwrite=True)
pipe_spy.add(upper_band, 'upper_band', overwrite=True)
results_spy_bol = run_pipeline(pipe_spy, start_date='2020-01-01', end_date='2020-05-01')
results_spy_bol = results_spy_bol.unstack()
results_spy_bol.columns = ['lower_band', 'middle', 'price', 'upper_band']
results_spy_bol
```

```python
results_spy_bol['lower_band'].plot(color='black', linestyle='-.', linewidth=3, figsize=(12, 6), legend='lower_band')
results_spy_bol['upper_band'].plot(color='black', linestyle=':', linewidth=3, figsize=(12, 6), legend='upper_band')
results_spy_bol['middle'].plot(color='black', linestyle='--', linewidth=1, figsize=(12, 6), legend='middle')
results_spy_bol['price'].plot(color='black', linestyle='-', linewidth=1, figsize=(12, 6), legend='price')
```

```python
# Filters
```

```python
sell_filter = USEquityPricing.close.latest > upper_band
sell_filter
```

```python
buy_filter = USEquityPricing.close.latest < lower_band
buy_filter
```

```python
pipe_spy.add(buy_filter, 'buy_filter', overwrite=True)
pipe_spy.add(sell_filter, 'sell_filter', overwrite=True)
results_spy_bol = run_pipeline(pipe_spy, start_date='2020-01-01', end_date='2020-05-01')
results_spy_bol = results_spy_bol.unstack()
results_spy_bol
```

```python
# Masking
```

```python
from quantopian.pipeline.factors import EWMA
ewma = EWMA(inputs=[USEquityPricing.close], window_length=20, decay_rate=5, mask=spy_etf)
ewma
```

```python
pipe_spy_ewma = Pipeline(columns={'ewma': ewma})
pipe_spy_ewma
```

```python
results_spy_ewma = run_pipeline(pipe_spy_ewma, start_date='2020-01-01', end_date='2020-05-01')
results_spy_ewma = results_spy_ewma.unstack()
results_spy_ewma
```

```python
pipe_spy_ewma = Pipeline(columns={'price': USEquityPricing.close.latest, 'ewma': ewma}, screen=spy_etf)
results_spy_ewma = run_pipeline(pipe_spy_ewma, start_date='2020-01-01', end_date='2020-05-01')
results_spy_ewma = results_spy_ewma.unstack()
results_spy_ewma.columns = ['ewma', 'price']
results_spy_ewma
```

```python
results_spy_ewma['price'].plot(color='black', linestyle='-', linewidth=1, figsize=(12, 6), legend='price')
results_spy_ewma['ewma'].plot(color='black', linestyle='--', linewidth=5, figsize=(12, 6), legend='ewma')
```

```python
# Classifiers
```

```python
from quantopian.pipeline.classifiers.morningstar import Sector
sector = Sector(mask=us_tradeable)
sector
```

```python
pipe_us_sector_ind = Pipeline(columns={'sector': sector}, screen=us_tradeable)
results_us_sector_ind = run_pipeline(pipe_us_sector_ind, start_date='2020-05-01', end_date='2020-05-01')
results_us_sector_ind
```

```python
SECTOR_CODE_MAP = {  
    101: 'Basic Materials',  
    102: 'Consumer Cyclical',  
    103: 'Financial Services',  
    104: 'Real Estate',  
    205: 'Consumer Defensive',  
    206: 'Healthcare',  
    207: 'Utilities',  
    308: 'Communication Services',  
    309: 'Energy',  
    310: 'Industrials',  
    311: 'Technology' ,  
}
results_us_sector_ind['sector'] = results_us_sector_ind['sector'].map(SECTOR_CODE_MAP)
results_us_sector_ind
```

```python
results_us_sector_ind['sector'].value_counts()
```

```python
from quantopian.pipeline.data import morningstar
mkt_cap = morningstar.valuation.market_cap.latest
mkt_cap
```

```python
us_tech = us_tradeable & ( sector.eq(310) | sector.eq(311) )
pipe_us_tech = Pipeline(columns={'sector': sector, 'volume': USEquityPricing.volume.latest, 'mkt_cap': mkt_cap/1e9}, screen=us_tech)
results_us_tech = run_pipeline(pipe_us_tech, start_date='2020-05-01', end_date='2020-05-01')
results_us_tech['sector'] = results_us_tech['sector'].map(SECTOR_CODE_MAP)
results_us_tech.sort_values(by='mkt_cap', ascending=False, inplace=True)
results_us_tech
```
